```python
#Question 6 — Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

import numpy as np
import matplotlib.pyplot as plt

sample =np.random.binomial(10, 0.5, 10000)
plt.hist(sample, bins=10, edgecolor='black')
plt.title('Histogram of Binomial Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

print("Mean:",sample.mean())
print("Variance:",sample.var())
```

```python
#Question 7 — Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

import numpy as np
from scipy.stats import norm

np.random.seed(123) # Set seed for reproducibility

sample_data = np.random.normal(loc=50, scale=1, size=40)

print("Generated Sample Data:", sample_data)
print("\nSample Size:", len(sample_data))

sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")

sigma_known = 1   # Assume known population standard deviation (example)
mu_0 = 50     # Null hypothesis mean

z_stat = (sample_mean - mu_0) / (sigma_known / np.sqrt(n))
p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))

print("=== Z-TEST RESULTS (Population σ Known) ===")
print(f"Z Statistic: {z_stat:.4f}")
print(f"P-value: {p_value_z:.6f}")

if p_value_z < 0.05:
    print("Result: Reject H₀ at α = 0.05")
else:
    print("Result: Fail to Reject H₀ at α = 0.05")

```

```python
from scipy.stats import ttest_1samp

t_stat, p_value_t = ttest_1samp(sample_data, mu_0)

print("=== T-TEST RESULTS (Population σ Unknown) ===")
print(f"T Statistic: {t_stat:.4f}")
print(f"P-value: {p_value_t:.6f}")

if p_value_t < 0.05:
    print("Result: Reject H₀ at α = 0.05")
else:
    print("Result: Fail to Reject H₀ at α = 0.05")

print("=== INTERPRETATION SUMMARY ===")
print(f"Sample mean = {sample_mean:.4f} for n = {n}")
print("Z-test is valid only if population σ is truly known.")
print("t-test is the correct test when σ is estimated from the sample.")
```

```python
#Question 8 — Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# creating sample data from a normal distribution
np.random.seed(10)
data = np.random.normal(50, 5, 200)   # mean=50, sd=5, sample size=200

# basic statistics
n = len(data)
mean_val = np.mean(data)
std_val = np.std(data, ddof=1)
se = std_val / np.sqrt(n)

# 95% confidence interval using t distribution
t_value = t.ppf(0.975, df=n-1)
lower = mean_val - t_value * se
upper = mean_val + t_value * se

print("Sample Mean:", mean_val)
print("Sample Std Dev:", std_val)
print("Standard Error:", se)
print("95% Confidence Interval:", (lower, upper))

# plotting histogram
plt.hist(data, bins=20)
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
#Question 9 — Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

# sample data
np.random.seed(5)
data = np.random.normal(100, 15, 40)

# calculating z-scores
mean_val = np.mean(data)
std_val = np.std(data, ddof=1)

z_scores = (data - mean_val) / std_val

print("Mean of data:", mean_val)
print("Standard deviation:", std_val)
print("First 10 z-scores:", z_scores[:10])

# histogram of z-scores
plt.hist(z_scores, bins=10)
plt.title("Histogram of Z-Scores")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.show()
```
