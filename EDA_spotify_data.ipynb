```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import warnings
warnings.filterwarnings("ignore")

print("Libraries imported successfully.")
```

```python
spotify_df = pd.read_csv("/content/sample_data/spotify.csv")
spotify_df.head()
```

```python
spotify_df.info()
```

```python
duplicate_rows = spotify_df[spotify_df.duplicated(['Track ID', 'Track Name'])]
if duplicate_rows.empty:
    print("No duplicate tracks found based on 'Track ID' and 'Track Name'.")
else:
    print(f"Found {len(duplicate_rows)} duplicate tracks:")
    display(duplicate_rows)
```

```python
initial_rows = len(spotify_df)
spotify_df.drop_duplicates(subset=['Track ID', 'Track Name'], inplace=True)
spotify_df.reset_index(drop=True, inplace=True)
final_rows = len(spotify_df)
print(f"Removed {initial_rows - final_rows} duplicate rows.")
print(f"New number of rows in spotify_df: {final_rows}")
```

```python
duplicate_rows_after_removal = spotify_df[spotify_df.duplicated(['Track ID', 'Track Name'])]
if duplicate_rows_after_removal.empty:
    print("No duplicate tracks found after removal.")
else:
    print(f"Found {len(duplicate_rows_after_removal)} duplicate tracks after removal:")
    display(duplicate_rows_after_removal)
```

```python
# What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.

# Calculate descriptive statistics for 'Popularity'
popularity_stats = spotify_df['Popularity'].describe()
print("Descriptive Statistics for Track Popularity Scores:")
print(popularity_stats)

# Visualize the distribution using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(spotify_df['Popularity'], bins=20, kde=True, color='purple')
plt.title('Distribution of Track Popularity Scores')
plt.xlabel('Popularity Score')
plt.ylabel('Number of Tracks')
plt.grid(True)
plt.show()

# Insights:
# - The distribution is somewhat left-skewed, indicating a longer tail towards lower popularity scores, but most tracks are concentrated in higher popularity ranges.
# - Most tracks fall within the 70 to 85 popularity range, suggesting the dataset primarily contains moderately to highly popular songs.
# - There are fewer tracks with very low popularity (below 50) and a few tracks that achieve very high popularity (above 90).
# - In conclusion, the dataset is weighted towards more popular tracks, with a significant number of songs having popularity scores above the average.
```

```python
# Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.

# Calculate the Pearson correlation coefficient
correlation = spotify_df['Popularity'].corr(spotify_df['Duration (ms)'])
print(f"Pearson correlation between Popularity and Duration (ms): {correlation:.2f}")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration (ms)', y='Popularity', data=spotify_df, alpha=0.6)
plt.title('Relationship between Track Duration and Popularity')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Insights:
# - The Pearson correlation coefficient (approximately 0.04) is very close to zero, indicating an extremely weak positive linear relationship.
# - The scatter plot visually confirms this. The points are widely scattered across the plot, showing no clear upward or downward trend.
# - There is no discernible pattern where tracks of a certain duration consistently have higher or lower popularity.
# - In conclusion, based on this dataset, there is no significant linear relationship or clear correlation between a track's duration and its popularity.
```
