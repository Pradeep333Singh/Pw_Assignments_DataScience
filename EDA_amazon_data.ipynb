```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print("Libraries imported successfully.")
```

```python
#Read the data
amazon_df = pd.read_csv('amazon.csv')
print("Dataset loaded successfully.")
print("First 5 rows of the dataset:")
amazon_df.head()
```

```python
amazon_df.describe(include='all')
```

```python
#Data cleaning

# 1. Convert 'discounted_price' and 'actual_price' to numeric
amazon_df['discounted_price'] = amazon_df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
amazon_df['actual_price'] = amazon_df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)

# 2. Convert 'discount_percentage' to numeric
amazon_df['discount_percentage'] = amazon_df['discount_percentage'].astype(str).str.replace('%', '').str.replace(',', '').astype(float)

# 3. Convert 'rating' to numeric, handling non-numeric values robustly
amazon_df['rating'] = amazon_df['rating'].replace({'No rating available': np.nan, '5 out of 5 stars': '5.0'})
amazon_df['rating'] = pd.to_numeric(amazon_df['rating'], errors='coerce')

# 4. Convert 'rating_count' to numeric
amazon_df['rating_count'] = amazon_df['rating_count'].astype(str).str.replace(',', '').astype(float)

# 5. Create 'primary_category' column
amazon_df['primary_category'] = amazon_df['category'].apply(lambda x: x.split('|')[0] if pd.notna(x) else np.nan)

# 6. Drop rows with NaN values in critical columns
critical_columns = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count', 'primary_category']
amazon_df.dropna(subset=critical_columns, inplace=True)

print("Data cleaning steps completed successfully.")
print("Info after cleaning:")
amazon_df.info()
```

```python
print("Descriptive statistics of the amazon_df DataFrame after cleaning:")
amazon_df.describe(include='all')
```

```python
# Question 1: Average Rating by Category
average_rating_by_category = amazon_df.groupby('primary_category')['rating'].mean()
print("Average rating by category:")
print(average_rating_by_category)
```

```python
#Question 2: Top Rating Count Products by Category
top_rated_products_by_category = amazon_df.loc[amazon_df.groupby('primary_category')['rating_count'].idxmax()]
print("Top products by rating count per category:")
print(top_rated_products_by_category[['primary_category', 'product_name', 'rating_count']])
```

```python
#Question 3: Distribution of Discounted Prices Versus Actual Prices
plt.figure(figsize=(12, 6))
sns.kdeplot(amazon_df['actual_price'], label='Actual Price', color='skyblue', fill=True)
sns.kdeplot(amazon_df['discounted_price'], label='Discounted Price', color='salmon', fill=True)
plt.title('Distribution of Actual vs. Discounted Prices')
plt.xlabel('Price (₹)')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#Question 4: Average Discount Percentage by Category
average_discount_by_category = amazon_df.groupby('primary_category')['discount_percentage'].mean()
print("Average discount percentage by category:")
print(average_discount_by_category)
```

```python
#Question 5: Most Popular Product Names
popular_product_names = amazon_df['product_name'].value_counts().head(10)
print("Most popular product names:")
print(popular_product_names)
```

```python
#Question 6: Most Popular Product Keywords from Names and Descriptions
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

# Download necessary NLTK data (if not already downloaded)
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords', quiet=True)

try:
    word_tokenize('test sentence')
except LookupError:
    print("NLTK 'punkt' resource (or its dependencies like 'punkt_tab') not found or incomplete.")
    print("Attempting to download 'punkt_tab' specifically, as suggested by the NLTK error message.")
    # The traceback explicitly suggested downloading 'punkt_tab'
    nltk.download('punkt_tab', quiet=True)

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = str(text).lower() # Convert to string and lowercase
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text) # Remove punctuation
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word not in stop_words and len(word) > 2] # Remove stopwords and short words
    return filtered_tokens

# Apply preprocessing to product_name and about_product
amazon_df['product_name_keywords'] = amazon_df['product_name'].apply(preprocess_text)
amazon_df['about_product_keywords'] = amazon_df['about_product'].apply(preprocess_text)

# Combine all keywords into a single list and count frequency
all_keywords = [word for sublist in amazon_df['product_name_keywords'] for word in sublist] + \
               [word for sublist in amazon_df['about_product_keywords'] for word in sublist]

from collections import Counter
popular_keywords = Counter(all_keywords).most_common(20)

print("Most popular keywords from product names and descriptions:")
for keyword, count in popular_keywords:
    print(f"{keyword}: {count}")
```

```python
#Question 7: Most Popular Product Reviews
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
from collections import Counter

# Ensure NLTK data is downloaded (if not already downloaded) from previous steps
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords', quiet=True)

try:
    word_tokenize('test sentence')
except LookupError:
    # This block should ideally not be reached if previous step fixed NLTK downloads
    print("NLTK 'punkt' resource not found or incomplete. Attempting comprehensive download.")
    nltk.download('punkt', quiet=True)
    nltk.download('punkt_tab', quiet=True) # Explicitly download punkt_tab if needed

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = str(text).lower() # Convert to string and lowercase
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text) # Remove punctuation
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word not in stop_words and len(word) > 2] # Remove stopwords and short words
    return filtered_tokens

# Apply preprocessing to review_content
amazon_df['review_keywords'] = amazon_df['review_content'].apply(preprocess_text)

# Combine all review keywords into a single list
all_review_keywords = [word for sublist in amazon_df['review_keywords'] for word in sublist]

# Count frequency of each keyword
popular_review_keywords = Counter(all_review_keywords).most_common(20)

print("Most popular keywords from product reviews:")
for keyword, count in popular_review_keywords:
    print(f"{keyword}: {count}")
```

```python
#Question 8: Correlation between Discounted Price and Product Rating
correlation = amazon_df['discounted_price'].corr(amazon_df['rating'])
print(f"Correlation between Discounted Price and Product Rating: {correlation:.4f}")
```

```python
#Ques 9: Average rating for each primary category
top_5_categories_by_rating = amazon_df.groupby('primary_category')['rating'].mean().nlargest(5)
print("Top 5 categories by average rating:")
print(top_5_categories_by_rating)
```
