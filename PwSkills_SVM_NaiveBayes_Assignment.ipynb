## Q1. What is a Support Vector Machine (SVM)?

**Answer:**

A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. It works by finding an optimal hyperplane that maximizes the margin between different classes.

## Q2. What is the difference between Hard Margin and Soft Margin SVM?

**Answer:**

Hard Margin SVM assumes data is perfectly separable and does not allow misclassification, whereas Soft Margin SVM allows some misclassification using a regularization parameter C to handle noisy data.

## Q3. What is the mathematical intuition behind SVM?

**Answer:**

SVM aims to maximize the margin between classes by minimizing the norm of the weight vector while satisfying classification constraints.

## Q4. What is the role of Lagrange Multipliers in SVM?

**Answer:**

Lagrange multipliers transform the constrained optimization problem into a dual problem, enabling kernel usage and identification of support vectors.

## Q5. What are Support Vectors in SVM?

**Answer:**

Support vectors are data points closest to the decision boundary and directly influence the position of the hyperplane.

## Q6. What is a Support Vector Classifier (SVC)?

**Answer:**

Support Vector Classifier is the classification version of SVM used for categorical target variables.

## Q7. What is a Support Vector Regressor (SVR)?

**Answer:**

SVR is the regression version of SVM that fits data within an epsilon-insensitive tube.

## Q8. What is the Kernel Trick in SVM?

**Answer:**

The kernel trick allows SVM to perform non-linear classification by implicitly mapping data into higher dimensions.

## Q9. Compare Linear, Polynomial and RBF kernels.

**Answer:**

Linear kernel is used for linearly separable data, Polynomial kernel captures feature interactions, and RBF kernel handles complex non-linear relationships.

## Q10. What is the effect of C parameter in SVM?

**Answer:**

C controls the trade-off between margin width and misclassification. High C leads to low bias and high variance.

## Q11. What is the role of Gamma in RBF kernel?

**Answer:**

Gamma defines the influence of individual data points. High gamma causes overfitting, low gamma causes underfitting.

## Q12. What is Naïve Bayes and why is it called Naïve?

**Answer:**

Naïve Bayes is a probabilistic classifier based on Bayes’ theorem with an assumption of feature independence.

## Q13. What is Bayes’ Theorem?

**Answer:**

Bayes’ theorem calculates conditional probability using prior knowledge.

## Q14. Difference between Gaussian, Multinomial and Bernoulli Naïve Bayes.

**Answer:**

Gaussian is used for continuous data, Multinomial for count data, and Bernoulli for binary features.

## Q15. When should Gaussian Naïve Bayes be used?

**Answer:**

When features are continuous and approximately normally distributed.

## Q16. What are the key assumptions of Naïve Bayes?

**Answer:**

Features are independent, equally important, and correctly represented by probability distributions.

## Q17. Advantages and disadvantages of Naïve Bayes.

**Answer:**

Advantages include simplicity and speed; disadvantages include unrealistic independence assumptions.

## Q18. Why is Naïve Bayes good for text classification?

**Answer:**

It handles high-dimensional sparse data efficiently and performs well with limited training data.

## Q19. Compare SVM and Naïve Bayes.

**Answer:**

SVM provides higher accuracy but is computationally expensive, whereas Naïve Bayes is faster and scalable.

## Q20. How does Laplace Smoothing help in Naïve Bayes?

**Answer:**

Laplace smoothing prevents zero probabilities by adding a constant to frequency counts.

## Q26. Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20 Newsgroups dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# 1. Load the Dataset
# We select 4 specific categories for faster execution and clearer visualization.
# To use the full dataset, remove the 'categories' argument.
categories = ['alt.atheism', 'soc.religion.christian', 'comp.graphics', 'sci.med']

print("Loading data...")
train_data = fetch_20newsgroups(subset='train', categories=categories, shuffle=True, random_state=42, remove=('headers', 'footers', 'quotes'))
test_data = fetch_20newsgroups(subset='test', categories=categories, shuffle=True, random_state=42, remove=('headers', 'footers', 'quotes'))

# 2. Create the Pipeline (TF-IDF Vectorizer + Naive Bayes)
# TfidfVectorizer converts text to numerical features; MultinomialNB predicts the category.
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# 3. Train the Model
print("Training model...")
model.fit(train_data.data, train_data.target)

# 4. Evaluate the Model
predictions = model.predict(test_data.data)
accuracy = accuracy_score(test_data.target, predictions)

print(f"\nModel Accuracy: {accuracy:.2%}")
print("\nClassification Report:\n")
print(classification_report(test_data.target, predictions, target_names=test_data.target_names))

# 5. Visualize Confusion Matrix
plt.figure(figsize=(8, 6))
mat = confusion_matrix(test_data.target, predictions)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, cmap='Blues',
            xticklabels=train_data.target_names, yticklabels=train_data.target_names)
plt.xlabel('True Label')
plt.ylabel('Predicted Label')
plt.title('Confusion Matrix')
plt.show()

# 6. Test on a custom string
new_doc = "The doctor prescribed some medicine for the patient."
pred_idx = model.predict([new_doc])[0]
print(f"\nTest Sentence: '{new_doc}'")
print(f"Predicted Category: {train_data.target_names[pred_idx]}")
```

## Q27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs

# Generate synthetic data
X, y = make_blobs(n_samples=100, centers=2, random_state=6)

# Create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# C values to compare
C_values = [0.01, 1, 100]

plt.figure(figsize=(15, 5))

for i, C in enumerate(C_values):
    # Train SVM
    clf = svm.SVC(kernel='linear', C=C)
    clf.fit(X, y)

    # Predict on mesh
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Plot
    plt.subplot(1, 3, i + 1)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title(f'SVM with C={C}')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

plt.show()
```

## Q29. Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score

# Load dataset (features have different scales)
data = load_wine()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. Without Scaling
clf_unscaled = SVC(kernel='rbf', gamma='auto') # gamma='auto' is bad without scaling
clf_unscaled.fit(X_train, y_train)
acc_unscaled = clf_unscaled.score(X_test, y_test)

# 2. With Scaling
clf_scaled = make_pipeline(StandardScaler(), SVC(kernel='rbf', gamma='auto'))
clf_scaled.fit(X_train, y_train)
acc_scaled = clf_scaled.score(X_test, y_test)

print(f"Accuracy without scaling: {acc_unscaled:.2f}")
print(f"Accuracy with scaling:    {acc_scaled:.2f}")
```

## Q30. Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and after Laplace Smoothing

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load data
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. Standard Gaussian NB (Default var_smoothing=1e-9)
gnb_default = GaussianNB()
gnb_default.fit(X_train, y_train)
y_pred_def = gnb_default.predict(X_test)

# 2. Gaussian NB with High Smoothing (var_smoothing=1e-1)
# This effectively widens the Gaussian curve, reducing overfitting risk but potentially underfitting
gnb_smooth = GaussianNB(var_smoothing=1e-1)
gnb_smooth.fit(X_train, y_train)
y_pred_smooth = gnb_smooth.predict(X_test)

print(f"Accuracy (Default Smoothing): {accuracy_score(y_test, y_pred_def):.2f}")
print(f"Accuracy (High Smoothing):    {accuracy_score(y_test, y_pred_smooth):.2f}")
```

## Q31. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C,gamma, kernel)

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.datasets import load_iris

# Load Data
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'poly', 'sigmoid']
}

# Run Grid Search
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=1, cv=3)
grid.fit(X_train, y_train)

# Print best results
print(f"Best Parameters: {grid.best_params_}")
print(f"Best Estimator Accuracy: {grid.best_estimator_.score(X_test, y_test):.2f}")
```

## Q32. Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy

```python
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Generate Imbalanced Data (95% class 0, 5% class 1)
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2,
                           weights=[0.95, 0.05], random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. Standard SVM
clf_std = SVC(kernel='linear')
clf_std.fit(X_train, y_train)
print("--- Standard SVM ---")
print(classification_report(y_test, clf_std.predict(X_test), target_names=['Class 0', 'Class 1']))

# 2. Weighted SVM
clf_weighted = SVC(kernel='linear', class_weight='balanced')
clf_weighted.fit(X_train, y_train)
print("\n--- Weighted SVM ---")
print(classification_report(y_test, clf_weighted.predict(X_test), target_names=['Class 0', 'Class 1']))
```

## Q33. Write a Python program to implement a Naïve Bayes classifier for spam detection using email data

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Sample Data
emails = [
    "Win a free lottery ticket now",
    "Meeting regarding the project update",
    "Cheap loans available apply today",
    "Family dinner plans for tonight",
    "Exclusive offer just for you",
    "Please review the attached document"
]
labels = [1, 0, 1, 0, 1, 0] # 1=Spam, 0=Ham

# Split and Vectorize
X_train, X_test, y_train, y_test = train_test_split(emails, labels, test_size=0.3, random_state=42)
vectorizer = CountVectorizer()
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)

# Train Multinomial NB
clf = MultinomialNB()
clf.fit(X_train_counts, y_train)

# Predict
y_pred = clf.predict(X_test_counts)
print(f"Predictions: {y_pred} (1=Spam, 0=Ham)")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

## Q35. Write a Python program to perform feature selection before training a Naïve Bayes classifier and compare results

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.naive_bayes import MultinomialNB
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score

# Load smaller subset of data
cats = ['alt.atheism', 'sci.space']
data_train = fetch_20newsgroups(subset='train', categories=cats)
data_test = fetch_20newsgroups(subset='test', categories=cats)

# 1. Without Feature Selection
model_full = make_pipeline(CountVectorizer(), MultinomialNB())
model_full.fit(data_train.data, data_train.target)
acc_full = accuracy_score(data_test.target, model_full.predict(data_test.data))

# 2. With Feature Selection (Top 500 features)
model_select = make_pipeline(
    CountVectorizer(),
    SelectKBest(chi2, k=500),
    MultinomialNB()
)
model_select.fit(data_train.data, data_train.target)
acc_select = accuracy_score(data_test.target, model_select.predict(data_test.data))

print(f"Accuracy (All Features):    {acc_full:.4f}")
print(f"Accuracy (Top 500 Features): {acc_select:.4f}")
```

## Q36.  Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO) strategies on the Wine dataset and compare their accuracy

```python
from sklearn.svm import SVC
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load Data
X, y = load_wine(return_X_y=True)
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. One-vs-Rest (ovr)
clf_ovr = SVC(decision_function_shape='ovr', kernel='rbf')
clf_ovr.fit(X_train, y_train)
acc_ovr = accuracy_score(y_test, clf_ovr.predict(X_test))

# 2. One-vs-One (ovo)
clf_ovo = SVC(decision_function_shape='ovo', kernel='rbf')
clf_ovo.fit(X_train, y_train)
acc_ovo = accuracy_score(y_test, clf_ovo.predict(X_test))

print(f"Accuracy OvR: {acc_ovr:.4f}")
print(f"Accuracy OvO: {acc_ovo:.4f}")
```

## Q39.  Write a Python program to train a Naïve Bayes classifier using different prior probabilities and compare performance

## Q40.  Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier andcompare accuracy

```python
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load Digits dataset
data = load_digits()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Linear SVM is required for RFE to access coef_ or feature_importances_
svc = SVC(kernel="linear", C=1)

# Select top 20 features (out of 64)
rfe = RFE(estimator=svc, n_features_to_select=20, step=1)
rfe.fit(X_train, y_train)

y_pred = rfe.predict(X_test)
print(f"Accuracy with Top 20 Features: {accuracy_score(y_test, y_pred):.4f}")
print(f"Number of Features used: {rfe.n_features_}")
```

## Q41. Write a Python program to train an SVM Classifier and evaluate its performance using Precision, Recall, and F1-Score instead of accuracy

```python
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, precision_recall_fscore_support

X, y = make_classification(n_samples=500, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

clf = SVC(kernel='linear')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# Detailed Report
print(classification_report(y_test, y_pred))

# Get raw values
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary')
print(f"Precision: {precision:.2f}, Recall: {recall:.2f}, F1: {f1:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler

data = load_wine()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale and Train
clf = SVC(kernel='linear', C=1).fit(StandardScaler().fit_transform(X_train), y_train)
y_pred = clf.predict(StandardScaler().fit(X_train).transform(X_test))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=data.target_names, yticklabels=data.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('SVM Confusion Matrix')
plt.show()
```

## Q44. Write a Python program to train an SVM Regressor (SVR) and evaluate its performance using Mean Absolute Error (MAE) instead of MSE

```python
from sklearn.svm import SVR
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error

# Load Housing Data (Regression task)
data = fetch_california_housing()
# Use small subset for speed
X, y = data.data[:1000], data.target[:1000]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale Data (Critical for SVR)
scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

# Train SVR
regressor = SVR(kernel='rbf', C=1.0, epsilon=0.2)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, y_pred):.4f}")
```

## Q45. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

X, y = make_classification(n_samples=1000, n_classes=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = GaussianNB()
clf.fit(X_train, y_train)

# Predict Probabilities (Positive class)
y_probs = clf.predict_proba(X_test)[:, 1]

# Calculate Score
auc = roc_auc_score(y_test, y_probs)
print(f"ROC-AUC Score: {auc:.4f}")

# Plot
fpr, tpr, _ = roc_curve(y_test, y_probs)
plt.plot(fpr, tpr, label=f"AUC={auc:.2f}")
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

## Q46.  Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import PrecisionRecallDisplay
import matplotlib.pyplot as plt

# Generate dataset
X, y = make_classification(n_samples=1000, n_classes=2, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train SVM (Enable probability for curves if needed, or use decision_function)
clf = SVC(kernel='linear', C=1)
clf.fit(X_train, y_train)

# Plot PR Curve
disp = PrecisionRecallDisplay.from_estimator(clf, X_test, y_test)
disp.ax_.set_title('Precision-Recall Curve')
plt.show()
```
